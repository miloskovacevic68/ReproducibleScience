#### Ćelije sa kodom u Pajtonu

Ćelije se **podrazumevano** tretiraju kao <span style="color:red">kodne</span> (kod u Pajtonu) i mogu se **izvršiti**, pri čemu se prikaz rada programa iz ćelije ispisuje neposredno ispod.

Na primer, treba prikazati kod funkcije koja vraća listu brojeva u opsegu od $a$ do $b$ koji sadrže cifru $c$ u svom zapisu, pa potom i rad funkcije na nekom primeru:


```python
# Primer koda u Pajtonu
def brojevi(a, b, c):
    """Vraća listu brojeva od a do b koji imaju cifru c u svom zapisu"""
    return [n for n in range(a, b+1) if str(c) in str(n)]

brojevi(10, 100, 7) # nije neophodna funkcija print, 
                    # Jupyter zna da pogodno formatira i prikaže rezultat 
```

Ništa nas ne srečava da, prilikom izlaza, koristimo i grafički prikaz. Na primer, treba nacrtati funkciju $f(x, y) = \sin{\sqrt{x^2 + y^2}}$  

```python
%matplotlib inline
# omogućava prikaz grafikona unutar ćelije a ne u prozoru

import matplotlib.pyplot as plt
import numpy as np

# definise interval za x i y i tačke u ravni za koje se računa f(x, y)
x, y = np.linspace(-5, 5, 100), np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# formira matricu vrednosti funkcije po svim
Z = np.sin(np.sqrt(X**2 + Y**2))

# crta 3d površinski grafik
tekuci = plt.subplot(projection="3d")
plt.title("z = sin((x^2 + y^2)^0.5)")
tekuci.plot_surface(X, Y, Z, cmap="winter")
plt.show()
```
